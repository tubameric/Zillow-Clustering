```python
# Code snippet 1
# Importing data from the webserver
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/nvamsimohan/DallasDSA/main/Zillow%20Tarrant%20County%20Raw%20Data.csv')

data.shape
```

```python
# Code snippet 2
# Displaying the data columns
data.columns
```

```python
# Code snippet 3
# Importing anomaly detection algos from pycaret
from pycaret.anomaly import *
```

```python
# Code snippet 4
# Setting up the pycaret anomaly detection algo configuration
anomaly_setup = setup(data)
```

```python
# Code snippet 5
# Building an anomaly detection model using knn algo
knn_model = create_model('knn')
```

```python
# Code snippet 6
# Identifying/predicting the anomalies using the knn model
knn_anomalies= predict_model(knn_model,data)
```

```python
# Code snippet 7
# Exporting the anomaly scores file to a csv
knn_anomalies.to_csv("Zillow Tarrant County Anomalies.csv")
```

```python
# Code snippet 8
# Listing the anomalies
data_anomalies = knn_anomalies[knn_anomalies['Anomaly']==1]
data_anomalies.shape
data_anomalies
```

```python
# Code snippet 9
# Assigning the non-anomalies to a clean dataset
clean_no_anomalies = knn_anomalies[knn_anomalies['Anomaly']==0]

clean_no_anomalies.shape
```

```python
# Code snippet 10
# Importing clustering algorithms
from pycaret.clustering import *
```

```python
# Code snippet 10.1
# Setting up the pycaret anomaly detection algo configuration
cluster_setup =setup(clean_no_anomalies,session_id=123)
```

```python
# Code snippet 11
# Building a clustering model using kmeans algo
kmeans_model = create_model('kmeans')
```

```python
# Code snippet 12
# Identifying/predicting the clusters using the kmeans algo
kmeans_data = assign_model(kmeans_model)
```

```python
# Code snippet 13
# Displaying the data with clusters
kmeans_data

# Exporting to a csv file
kmeans_data.to_csv("Zillow Tarrant Data Clusters.csv")
```
